```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline


resultado = pd.read_csv("Respuestas Diagnostic survey for a tool used in astronomy.csv", parse_dates=['Timestamp'], sep = ';')
resultado2 = pd.read_csv("Diagnostic survey for a tool used in astronomy (Responses)2.csv", sep = ';')

```

```python
order = [
    'Yes',
    'No'
]
sns.countplot(y='Do you know what Astroquery is?',
               data=resultado, order=order,
              color='purple').set(title="Do you know what Astroquery is?",
                                ylabel="");
sns.despine()
```

```python
sns.countplot(y='What do you use Astroquery for?',
              data=resultado2,
              color='purple').set(title="What do you use Astroquery for?",
             ylabel="");
sns.despine()

    
```
